## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Research question:** Among a client's live pages, which ones should a content team review first for a refresh, given limited weekly capacity?

**Unit of analysis:** one content page (one row per `content_id`).

**Output:** a ranked review queue — each page gets a refresh-opportunity score plus 1-2 reason codes (e.g. *stale_visible_page*, *declining_with_demand*, *low_ctr_visible_page*) that tell the reviewer why it surfaced.

**The decision it improves:** how a small SEO/content team allocates its scarce review hours each week — turning "where do we even start?" into a prioritized shortlist.

**Who acts on it:** a content strategist or SEO reviewer who opens the top-N pages and decides whether to refresh, consolidate, or leave them.

**Cost of a wrong call:** a false positive wastes a reviewer's limited time on a page that did not need attention (opportunity cost — a genuinely at-risk page waits longer). A false negative means a page with real demand keeps decaying unnoticed and loses traffic. Because this is decision-support (not an automated edit), a wrong score misdirects attention rather than breaking anything — but attention is the scarce resource, so ranking quality is what matters.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd, glob
_hits = glob.glob("**/content_refresh_anonymized.csv", recursive=True)
_url = "https://raw.githubusercontent.com/Noorulhuda-12/flyrank/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(_hits[0]) if _hits else pd.read_csv(_url)
print("Total pages in starter slice:", len(df))
declining = (df["trend_direction"] == "down").mean()
print("Share of pages trending DOWN: {:.1%}".format(declining))
stale_visible = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).mean()
print("Share stale (>=180d) AND visible (>=500 impr) = prime refresh candidates: {:.1%}".format(stale_visible))
low_ctr = ((df["impressions_90d"] >= 500) & (df["avg_position"].between(1,20)) & (df["ctr"] < 0.5)).mean()
print("Share visible, top-20 position, CTR < 0.5% = CTR opportunity: {:.1%}".format(low_ctr))
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

**What this work *can* say:**
- *Observed / measured:* on the starter slice, a large share of pages are trending down and many are visible top-20 pages with low CTR — so there is a real pool of pages that plausibly warrant review (see numbers above).
- *Directional:* certain observable signals (staleness, declining trend with demand, low CTR at good positions) are associated with pages worth reviewing first.
- *Decision-support:* I can produce a ranked, reason-coded queue that helps a reviewer spend limited time on the most promising candidates first, and I can show (via client-holdout validation) whether a learned ranking beats a transparent baseline rule.

**What this work can *never* say:**
- It cannot prove that refreshing a page *causes* recovery — that needs an experiment or causal design this observational data cannot provide.
- It does not "predict Google's algorithm" or reveal any ranking formula.
- It is not a guarantee about any individual page; scores are prioritization signals, not certainties.
- All results are specific to this pseudonymized snapshot and must be re-earned with proper validation on the full warehouse.
